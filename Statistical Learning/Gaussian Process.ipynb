```python
import jax
import jax.numpy as jnp
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
jax.config.update("jax_enable_x64", True)
```

```python
# Squared Exponential Kernel
from tinygp import kernels

kernel = kernels.ExpSquared()
dx = np.linspace(0,5,100)
plt.plot(dx, kernel(dx, dx[:1]))
plt.xlabel("dx")
plt.ylabel("k(dx)")
plt.show()
```

```python
# Gaussian Process
from tinygp import GaussianProcess

X = np.sort(np.random.default_rng(1).uniform(0, 10, 100))

gp = GaussianProcess(kernel, X) # X
```

```python
y = gp.sample(jax.random.PRNGKey(3), shape=(4,))
plt.plot(X, y.T)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
