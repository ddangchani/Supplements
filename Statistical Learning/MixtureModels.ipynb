## Gaussian Mixture Model

```python
import numpy as np
import numpy.random
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
# SETUP
N = 10000
PARAMS = np.array([[-10, 5],
                   [1, 3],
                   [10, 3]])
N_COMPONENTS = PARAMS.shape[0]

WEIGHTS = np.ones(N_COMPONENTS, dtype=np.float64) / np.float64(N_COMPONENTS)
```

```python
# Generate random sample from Gaussian
mixture_idx = np.random.choice(N_COMPONENTS, N, replace=True, p=WEIGHTS)

Y = np.fromiter(
    (ss.norm.rvs(*(PARAMS[i])) for i in mixture_idx),
    dtype=np.float64
)
```

```python
# Histogram & PDF Plot

xs = np.linspace(Y.min(), Y.max(), 200)
ys = np.zeros_like(xs)

for (l, s), w in zip(PARAMS, WEIGHTS):
    ys += ss.norm.pdf(xs, loc=l, scale=s) * w

plt.plot(xs, ys, color='navy')
plt.hist(Y, bins='fd', color='skyblue', density=True)
plt.xlabel('X')
plt.ylabel('pdf(x)')
plt.title('Gaussian Mixture Model')
plt.show()
```
