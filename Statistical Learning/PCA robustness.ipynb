```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2 dimensional data

# 1. Generate data
np.random.seed(1)
x = np.random.randn(100, 2)
```

```python
# PCA

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(x)

plt.scatter(x[:, 0], x[:, 1])
plt.plot([0, 2 * pca.components_[0, 0]], [0, 2 * pca.components_[0, 1]], 'b', label='PC1')
plt.plot([0, 2 * pca.components_[1, 0]], [0, 2 * pca.components_[1, 1]], 'r', label='PC2')
plt.legend()
plt.savefig('plots/PCA_no_outlier.png', dpi=300, facecolor='white')
```

```python
# Add an outlier

x_outlier = np.copy(x)
x_outlier[0, :] = [5, 5]

# PCA
pca_outlier = PCA(n_components=2)
pca_outlier.fit(x_outlier)

plt.scatter(x_outlier[1:, 0], x_outlier[1:, 1])
plt.scatter(x_outlier[0, 0], x_outlier[0, 1], c='r')

plt.plot([0, 2 * pca_outlier.components_[0, 0]], [0, 2 * pca_outlier.components_[0, 1]], 'b', label='PC1')
plt.plot([0, 2 * pca_outlier.components_[1, 0]], [0, 2 * pca_outlier.components_[1, 1]], 'r', label='PC2')
plt.legend()
plt.savefig('plots/PCA_outlier.png', facecolor='white', dpi=300)
```
