# MH Algorithm Example Code
Sample From Normal distribution

```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
# Parameter
MU = 10.0
SIGMA = 5.0
target = ss.norm(loc=MU, scale=SIGMA) # target distribution
```

```python
# Rep count = 10000
x = np.zeros(shape=10000)

# initialize x_0
x[0] = 3.0

for i in range(1,10000):
    x_t = x[i-1]
    x_proposed = x_t + np.random.standard_normal(1)[0] # Proposal Distribution
    A = min(1, target.pdf(x_proposed) / target.pdf(x_t)) # Since q is symmetric

    u = np.random.uniform(size=1)[0] # u from Uniform dist

    if u <= A : # Accept
        x[i] = x_proposed
    else:
        x[i] = x_t
```

```python
# Histogram and pdf Plot
xs = np.linspace(x.min(), x.max(), 100)
ys = np.zeros_like(xs)
ys = target.pdf(xs)

plt.plot(xs, ys, color='navy')
plt.hist(x, bins='fd', color='skyblue', density=True)
plt.xlabel('X')
plt.ylabel('pdf(x)')
plt.title('Markov Chain Monte Carlo Sampling')
plt.show()
```
