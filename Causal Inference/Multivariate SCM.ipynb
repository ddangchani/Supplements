## 4-node SCM sample generation
#### Sample size(n) = 100
$$
\begin{aligned}
&f_1(x_3,n) = 2x_3+n\\
&f_2(x_1,n) = (0.5x_1)^2+n\\
&f_3(n) = n\\
&f_4(x_2,x_3,n) = x_2 + 2\sin(x_3+n)
\end{aligned}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# X3 먼저 생성(no parent node)
np.random.seed(3)
n_3 = np.random.randn(100) # noise variable from Normal
x_3 = n_3
x_3[:5]
```

```python
# X_1 생성(from X_3)
np.random.seed(1)
n_1 = np.random.randn(100)
x_1 = 2*x_3 + n_1
x_1[:5]
```

```python
# X_2 생성(from X_1)
np.random.seed(2)
n_2 = np.random.randn(100)
x_2 = np.square(0.5 * x_1) + n_2
x_2[:5]
```

```python
# X_4 생성(from X_2, X_3)
np.random.seed(3)
n_4 = np.random.randn(100)
x_4 = x_2 + 2 * np.sin(x_3 + n_4)
x_4[:5]
```

```python
# Plot between two 
figs, axes = plt.subplots(2,2,figsize = (10,10))
axes[0][0].scatter(x_3,x_1)
axes[0][0].set(xlabel = 'X_3', ylabel = 'X_1')
axes[0][1].scatter(x_1,x_2)
axes[0][1].set(xlabel = 'X_1', ylabel  = 'X_2')
axes[1][0].scatter(x_3,x_4)
axes[1][0].set(xlabel = 'X_3', ylabel  = 'X_4')
axes[1][1].scatter(x_2,x_4)
axes[1][1].set(xlabel = 'X_2', ylabel  = 'X_4')
plt.suptitle('Scatter between two components',fontsize=14)

plt.show()
```
