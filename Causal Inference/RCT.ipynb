```python
import pgmpy
from pgmpy.models import BayesianNetwork
from causalgraphicalmodels import CausalGraphicalModel
```

```python
model = BayesianNetwork(
    [('T','Y'),('X','T'),('X','Y')]
)
model2 = CausalGraphicalModel(model.nodes, model.edges)
model2.draw()
```
