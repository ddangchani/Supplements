```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
def f1(x):
    return 0.25 * x + 0.05

def f2(x):
    return x 

def f3(x):
    return 0.25 - 0.5*x
```

```python
# Plot the functions
fig, axes = plt.subplots(3,3, figsize=(12,8))

x = np.linspace(-1,1,100)

axes[0][0].plot(x, f1(x), label='f1', color='r')
axes[0][1].plot(x, f2(x), label='f2', color='g')
axes[0][2].plot(x, f3(x), label='f3', color='b')

axes[1][0].plot(x, relu(f1(x)), label='relu(f1)', color='r')
axes[1][1].plot(x, relu(f2(x)), label='relu(f2)', color='g')
axes[1][2].plot(x, relu(f3(x)), label='relu(f3)', color='b')

axes[2][0].plot(x, relu(f1(x)) * (-1), label='-relu(f1)', color='r')
axes[2][1].plot(x, relu(f2(x)) * (1.5), label='1.5relu(f2)', color='g')
axes[2][2].plot(x, relu(f3(x)) * (0.6), label='0.6relu(f3)', color='b')

for ax in axes.ravel():
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.plot(x, np.zeros_like(x), 'k--', alpha=0.5)
    ax.set_ylim(-1,1)
    ax.set_xlim(-1,1)
    ax.legend(fontsize=10)

plt.tight_layout()
plt.savefig('universalapproximation.png', dpi=300, facecolor='w')
```

```python
def f(x):
    return (-0.5) + (-1) * relu(f1(x)) + 1.5 * relu(f2(x)) + 0.6 * relu(f3(x))

# spline points
# x1 = point where relu(f1(x)) = 0
# x2 = point where relu(f2(x)) = 0
# x3 = point where relu(f3(x)) = 0

x1 = -0.2
x2 = 0
x3 = 0.5
```

```python
# Plot the function
fig, ax = plt.subplots(figsize=(6,4))

x = np.linspace(-1,1,100)

ax.plot(x, f(x), label='f', color='navy')

ax.set_xlabel('x')
ax.set_ylabel('y')

# ax.plot(x, np.zeros_like(x), 'k--', alpha=0.5)
ax.vlines(x1, -1, 1, color='r', linestyle='--', alpha=0.5)
ax.vlines(x2, -1, 1, color='g', linestyle='--', alpha=0.5)
ax.vlines(x3, -1, 1, color='b', linestyle='--', alpha=0.5)

ax.set_ylim(-1,1)
ax.set_xlim(-1,1)

ax.legend(fontsize=10)

plt.tight_layout()
plt.savefig('universalapproximation2.png', dpi=300, facecolor='w')
```
